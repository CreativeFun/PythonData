## Grid search

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33,
                                                    )
grid_params = {
    'n_estimators': [25, 50, 200, 500],
    'max_depth': [4, 10, 20, 30],
    'criterion': ['gini', 'entropy']
}
# grid_params = {
#     'n_estimators': [25],
#     'max_depth': [4],
#     'criterion': ['gini', 'entropy']
# }
```

```python
res = {
    'n_estimators': [],
    'max_depth': [],
    'criterion': [],
    'accuracy': []
}

for n in grid_params['n_estimators']:
  for depth in grid_params['max_depth']:
    for crit in grid_params['criterion']:
      clf = RandomForestClassifier(max_depth=depth, random_state=0, n_estimators=n, criterion=crit)
      clf.fit(X_train,y_train)
      y_pred = clf.predict(X_test)
      accuracy = accuracy_score(y_test, y_pred)

      # save results
      res['n_estimators'].append(n)
      res['max_depth'].append(depth)
      res['criterion'].append(crit)
      res['accuracy'].append(accuracy)
```

```python
best_index = res['accuracy'].index(max(res['accuracy']))
best_index
best_params = {
    'n_estimators': res['n_estimators'][best_index],
    'max_depth': res['max_depth'][best_index],
    'criterion': res['criterion'][best_index],
    'accuracy':  res['accuracy'][best_index]
}
best_params
```

# Deep Learning


```python
import tensorflow
from tensorflow import keras
import tensorflow as tf


model = keras.Sequential(
    [
      keras.layers.Dense(9,activation="ReLU"),
      keras.layers.Dense(64,activation="ReLU"),
      keras.layers.Dense(128,activation="ReLU"),
      keras.layers.Dense(32,activation="ReLU"),
      keras.layers.Dense(1,activation="sigmoid"),
    ]
)

def my_loss(y_true):
    squared_difference = tf.square(y_true - 2)
    return tf.reduce_mean(squared_difference, axis=-1)  # Note the `axis=-1`

model.compile(optimizer='adam',
              loss=my_loss(y_train),
              metrics=[tensorflow.keras.metrics.BinaryAccuracy()
                       ])

history = model.fit(X_train,y_train,epochs=512,batch_size=16)


```

```python
ypred = (model.predict(X_test) > 0.5).astype(int).reshape(-1) #sprowadzenie do wektora
print(ypred.shape)
print(y_test.shape)

```

```python
confusion = confusion_matrix(y_test, ypred)
print(confusion)
print(classification_report(y_test, ypred, target_names=['yes','no']))
```
